```python
penpubs = {
    "1991":1,
    "1994":1,
    "1997":3,
    "1998":3,
    "1999":2,
    "2001":2,
    "2002":1,
    "2003":3,
    "2004":4,
    "2005":2,
    "2006":9,
    "2007":2,
    "2008":6,
    "2009":5,
    "2010":9,
    "2011":4,
    "2012":11,
    "2013":13,
    "2014":5,
    "2015":5,
    "2016":9,
    "2017":13,
    "2018":11,
    "2019":13,
    "2020":8,
    "2021":7,
    "2022":2,
    "2023":6
}

penpubs_df = pd.DataFrame.from_dict(penpubs, orient='index')
penpubs_df.plot(kind="bar",figsize=(15, 8))
```
